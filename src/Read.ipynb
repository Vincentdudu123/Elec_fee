```python

import pandas as pd
import numpy as np
import plotly as py 

import seaborn as sns
import plotly.express as px
import cufflinks as cf
import matplotlib.pyplot as plt
#%matplotlib inline 

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df = pd.read_excel("Elec_fee.xlsx")

import matplotlib.dates as mdates
    
plt.tight_layout()
Dataset=[]
for i in range(len(df.columns)-1):
    df1 = df.loc[df[df.columns[i+1]].notna()][[df.columns[0],df.columns[i+1]]]
    Dataset.append(df1)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.set_xlabel("Month")
    ax.plot(df1['Month'], df1[df1.columns[1]], label=df1.columns[1])
    ax.legend()

    ax.xaxis.set_major_formatter(mdates.DateFormatter('%y-%m'))
    plt.gcf().autofmt_xdate()
    fig.savefig(str(i)+".png",bbox_inches="tight",dpi=400, transparent=False,facecolor='w')
#    plt.plot(x='Month',y=df1.columns[1],figsize=(12,9),subplots=True)


    

```

```python

pic1 = ['Average price [euro/MWh]', 'Highest price[euro/MWh]', 'Lowest price[euro/MWh]']
pic2 = ['Hours of negative price[h]', 'hours of overprice[h]']
pic3 = ['total production[TWh]', 'total consumption[TWh]']
pic4 = ['percentage of renewable[%]', 'highest renewable[%]', 'lowest renewable[%]']

pic = [pic1,pic2,pic3,pic4]


for i in pic:
    fig = plt.figure()
    ax = fig.add_subplot(111)
    for ii in i:
        index = df.columns.tolist().index(ii)-1
        ax.plot(Dataset[index]["Month"],Dataset[index][ii],label=ii)

    ax.legend()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%y-%m'))
    plt.gcf().autofmt_xdate()

    fig.savefig(str(i)[:10]+".png",bbox_inches="tight",dpi=400, transparent=False,facecolor='w')
        
```
